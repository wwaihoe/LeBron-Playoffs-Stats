```python
import pandas as pd
import numpy as np
import lxml
```

#Data Preparation

```python
url = "https://www.basketball-reference.com/players/j/jamesle01/gamelog-playoffs/"
dfs = pd.read_html(url)
print(len(dfs))
```

```python
#Data needed is located in index 7
dfs[7].head(40)
```

```python
df_bsc = dfs[7]
```

```python
url = "https://www.basketball-reference.com/players/j/jamesle01/gamelog-playoffs-advanced/"
dfs1 = pd.read_html(url)
print(len(dfs))
```

```python
df_adv = dfs1[0]
df_adv.head()
```

```python
print(df_bsc.columns)
print(df_adv.columns)
```

```python
#Concatenate basic and advanced stats dataframes
df = pd.concat([df_bsc, df_adv[["TS%", "eFG%", "ORB%", "DRB%", "TRB%", "AST%",
       "STL%", "BLK%", "TOV%", "USG%", "ORtg", "DRtg", "GmSc"]]], axis=1)
df.head()
```

```python
df.columns
```

```python
df.rename(columns={"2006 Playoffs": "Date"}, inplace=True)
```

```python
df["Rk"].values
```

```python
#Remove non-useful rows
df.dropna(subset=["Rk"],inplace = True)
df = df.loc[df["Rk"] != "Rk"]
print(df.shape)
df["Rk"].values
```

```python
#Check first unnamed column
df["Unnamed: 5"].head()
```

```python
#Column "Unnamed: 5" shows whether game was an away game or home game
df["Unnamed: 5"].fillna("0",inplace = True)
#Create new columns for Home and Away games
df["Away"] = 0
df["Home"] = 0
for i in range(0,len(df)):
    if df["Unnamed: 5"].iloc[i] == "@":
        df["Away"].iloc[i] = 1
    else:
        df["Home"].iloc[i] = 1
df.head(30)
```

```python
#Check second unnamed column
df["Unnamed: 8"].head()
```

```python
#"Unnamed: 8" shows the outcome of game and point differential
#Create new column for outcome of game
df["Outcome"] = "L"
for i in range(0,len(df)):
    if "W" in str(df["Unnamed: 8"].iloc[i]):
        df["Outcome"].iloc[i] = "W"
#Create new column for point differential of game
df["Point Diff"] = ""
for i in range(0,len(df)):
    temp_str = str(df["Unnamed: 8"].iloc[i]).replace("W", "")
    temp_str = temp_str.replace("L", "")
    temp_str = temp_str.replace("(", "")
    temp_str = temp_str.replace("+", "")
    temp_str = temp_str.replace(")", "")
    temp_str = temp_str.replace(" ", "")
    df["Point Diff"].iloc[i] = float(temp_str)
df.head()
```

```python
for col in df.columns:
    print(col + ": " + str(df[col].isnull().sum()))
```

```python
#Only nan values remaining are in 3P% and FT% of games where Lebron does not shoot any free throws or threes
#Replace 3P% and FT% with 0 in the games where LeBron does not shoot any free throws or threes
df.fillna(0, inplace = True)
for col in df.columns:
    print(col + ": " + str(df[col].isnull().sum()))
```

```python
#Remove "+" in +/-
for i in range(0,len(df)):
    temp_str = str(df["+/-"].iloc[i]).replace("+", "")
    df["+/-"].iloc[i] = float(temp_str)
df.head()
```

```python
df.dtypes
```

```python
nonnum_col = ["Date", "Series", "Tm", "Unnamed: 5", "Opp", "Unnamed: 8", "MP", "Outcome"]
num_col = [col for col in df.columns if col not in nonnum_col]
for c in num_col:
    df[num_col] = df[num_col].astype(float)
df.dtypes
```
